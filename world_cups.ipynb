```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = pd.read_csv('worldcup_fifa.csv')
dados
```

```python
dados['winner'] = dados['winner'].replace('West Germany','Germany')
dados['country_name'] = dados['country_name'].replace('West Germany','Germany')
```

**Seleções com mais títulos**

```python
campeos = dados['winner']
```

```python
times = list(campeos.value_counts().index.values)
titulos = list(campeos.value_counts().values)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=times, y=titulos, edgecolor='k', palette='Blues_r')
plt.xticks(rotation = 45)
plt.title('Títulos por País', size=15)
plt.xlabel('Países')
plt.ylabel('Títulos')
plt.show()
```
